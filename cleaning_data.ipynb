```python
import pandas as pd


markalar = ['FIAT', 'PEUGEOT', 'RENAULT', 'HYUNDAI', 'FORD', 'DACIA', 'TOYOTA',
            'VOLKSWAGEN', 'OPEL', 'NISSAN', 'CITROEN', 'KIA', 'SKODA', 'AUDI',
            'MERCEDES_BENZ', 'VOLVO', 'BMW', 'MINI', 'LAND ROVER', 'MAZDA',
            'MITSUBISHI', 'LADA', 'CHEVROLET', 'SEAT', 'IVECO', 'HONDA', 'JEEP',
            'PORSCHE', 'ISUZU', 'LANCIA', 'JAGUAR', 'ASTON MARTIN', 'SUBARU',
            'OTOKAR', 'ALFA ROMEO', 'LINCOLN', 'DODGE', 'CHRYSLER', 'CHERY',
            'SUZUKI', 'TATA', 'TOFAS']

# Markalara karÅŸÄ±lÄ±k gelen modelleri saklamak iÃ§in bir sÃ¶zlÃ¼k
marka_model_sozluk = {}

# Her marka iÃ§in modelleri bulma ve saklama
for marka in markalar:
    modeller = df[df['Marka'] == marka]['Model'].unique()
    marka_model_sozluk[marka] = modeller

# Sonucu inceleme
for marka, modeller in marka_model_sozluk.items():
    print(f"{marka}: {list(modeller)}")
```
